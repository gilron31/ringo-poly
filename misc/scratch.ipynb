```python
model = cp_model.CpModel()
VARS = ["x", "y", "z"]
DEG = 3
N_REGS = 2
N_STEPS = 3
multivar_poly_processor = MultiVarPolyProcessor(
    model, VARS, DEG, N_REGS, N_STEPS, [AddOp(), SubOp()], bound=10
)
x, y, z = multivar_poly_processor.get_basis()
multivar_poly_processor.find_code([x, y], [x + y])
multivar_poly_processor.solve()
multivar_poly_processor.assert_solution_valid([x, y], [x + y])
print(multivar_poly_processor.dump_solved_state())
```

```python
model = cp_model.CpModel()
VARS = ["x0", "x1", "y0", "y1"]
DEG = 2
N_REGS = 5
N_STEPS = 5
multivar_poly_processor = MultiVarPolyProcessor(
    model, VARS, DEG, N_REGS, N_STEPS, [AddOp(), SubOp(), MulOp()], bound=2
)
x0, x1, y0, y1 = multivar_poly_processor.get_basis()
multivar_poly_processor.find_code(
    [x0, y0, x1, y1], [x0 * y0, x1 * y1, x1 * y0 + x0 * y1]
)
multivar_poly_processor.model.add(multivar_poly_processor.op_selectors[0][2] == True)
multivar_poly_processor.model.add(multivar_poly_processor.op_selectors[1][2] == True)
multivar_poly_processor.model.add(multivar_poly_processor.op_selectors[2][2] == True)
multivar_poly_processor.model.add(multivar_poly_processor.op_selectors[3][2] == True)
multivar_poly_processor.model.add(multivar_poly_processor.op_selectors[4][0] == True)
multivar_poly_processor.model.add(multivar_poly_processor.dst_selectors[0][4] == True)
multivar_poly_processor.model.add(multivar_poly_processor.dst_selectors[1][0] == True)
multivar_poly_processor.model.add(multivar_poly_processor.dst_selectors[2][3] == True)
multivar_poly_processor.model.add(multivar_poly_processor.dst_selectors[3][2] == True)
multivar_poly_processor.model.add(multivar_poly_processor.dst_selectors[4][2] == True)
multivar_poly_processor.solve()
multivar_poly_processor.assert_solution_valid(
    [x0, y0, x1, y1], [x0 * y0, x1 * y1, x1 * y0 + x0 * y1]
)
print(multivar_poly_processor.dump_solved_state())
```
